С помощью функции linspace из библиотеки Numpy создайте массив t из 51 числа от 0 до 10 включительно.<br>
Создайте массив Numpy под названием f, содержащий косинусы элементов массива t.<br>
Постройте линейную диаграмму, используя массив t для координат по горизонтали,а массив f - для координат по вертикали. Линия графика должна быть зеленого цвета.<br>
Выведите название диаграммы - 'График f(t)'. Также добавьте названия для горизонтальной оси - 'Значения t' и для вертикальной - 'Значения f'.
Ограничьте график по оси x значениями 0.5 и 9.5, а по оси y - значениями -2.5 и 2.5.


```python
t = np.linspace(0, 10, num=51)
```

```python
f = np.cos(t)
```

```python
plt.plot(t, f, color='green')
plt.title('График f(t)')
plt.xlabel('Значения t')
plt.ylabel('Значения f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

Создайте массивы y1, y2, y3, y4 по следующим формулам:<br>
y1 = x**2<br>
y2 = 2 * x + 0.5<br>
y3 = -3 * x - 1.5<br>
y4 = sin(x)<br>
Используя функцию subplots модуля matplotlib.pyplot, создайте объект matplotlib.figure.Figure с названием fig и массив объектов Axes под названием ax,причем так, чтобы у вас было 4 отдельных графика в сетке, состоящей из двух строк и двух столбцов. В каждом графике массив x используется для координат по горизонтали.В левом верхнем графике для координат по вертикали используйте y1,в правом верхнем - y2, в левом нижнем - y3, в правом нижнем - y4. Дайте название графикам: 'График y1', 'График y2' и т.д.<br>
Для графика в левом верхнем углу установите границы по оси x от -5 до 5.<br>
Установите размеры фигуры 8 дюймов по горизонтали и 6 дюймов по вертикали.<br>
Вертикальные и горизонтальные зазоры между графиками должны составлять 0.3.


```python
x = np.linspace(-3, 3, num=51)
```

```python
y1 = x ** 2
y2 = 2 * x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
plt.subplots_adjust(hspace=0.3,wspace=0.3)
fig, ax = plt.subplots(nrows=2,ncols=2)
fig.set_size_inches(8, 6)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.set_xlim([-5, 5])
ax1.set_title('График y1')
ax1.plot(x, y1)
ax2.set_title('График y2')
ax2.plot(x, y2)
ax3.set_title('График y3')
ax3.plot(x, y3)
ax4.set_title('График y4')
plt.plot(x, y4)
plt.show()
```

Создайте легенду с двумя значениями: “Class 0” и “Class 1”. Гистограмма обычных транзакций должна быть серого цвета, а мошеннических - красного. Горизонтальной оси дайте название “V1”.

```python
import pandas as pd
```

```python
df = pd.read_csv('creditcard.csv')
```

```python
data = df.value_counts('Class')
```

```python
plt.style.use('fivethirtyeight')
data.plot(kind='bar')
plt.show()
```

```python
data.plot(kind='bar', log=True)
plt.show()
```

```python
V1_Class1 = df.loc[df['Class'] == 1, ['V1']]
V1_Class0 = df.loc[df['Class'] == 0, ['V1']]
```

```python
fig, ax = plt.subplots(ncols=2)
fig.set_size_inches(8, 6)
fig.subplots_adjust(wspace=0.3, hspace=0.3)
ax1, ax2 = ax
ax1.hist(V1_Class1, density=True, color='red', bins=20, alpha=0.5)
ax1.legend(['Class 1'])
ax1.set_xlabel('V1')
ax2.hist(V1_Class0, density=True, color='gray', bins=20, alpha=0.5)
ax2.legend(['Class 0'])
ax2.set_xlabel('V1')
plt.show()
```
